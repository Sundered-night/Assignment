```python
import random

board = ["-", "-", "-",
        "-", "-", "-",
        "-", "-", "-"]
    
current_player = "X"
current_game = True
winner = None


##To check the current state of the gameboard
def game_board(board):
    print(board[0] + " | " + board[1] + " | " + board[2])
    print("---------")
    print(board[3] + " | " + board[4] + " | " + board[5])
    print("---------")
    print(board[6] + " | " + board[7] + " | " + board[8])
    

def player_input(board):
    grid = int(input(" Select a grid from 0 to 8"))   
    if board[grid]== "-" :
        board[grid] = current_player
    else :
        print("There is already a person here at the board")
        
        
def row(board):
    global winner
    if board[0] == board[3] == board[6] and board[0] != "-":
        winner = board[0]
        return True
    elif board[1] == board[4] == board[7] and board[1] != "-":
        winner = board[1]
        return True
    elif board[2] == board[5] == board[8] and board[2] != "-":
        winner = board[2]
        return True
    return False
        
        
def column(board):
    global winner
    if board[0] == board[1] == board[2] and board[0] != "-":
        winner = board[0]
        return True
    elif board[3] == board[4] == board[5] and board[3] != "-":
        winner = board[3]
        return True
    elif board[6] == board[7] == board[8] and board[6] != "-":
        winner = board[6]
        return True
    return False
    
    
def diag(board):
    global winner
    if board[0] == board[4] == board[8] and board[0] != "-":
        winner = board[0]
        return True
    elif board[2] == board[4] == board[6] and board[4] != "-":
        winner = board[2]
        return True
    return False
    
    
def tie(board):
    global current_game
    if "-" not in board:
        game_board(board)
        print("It is a tie!")
        current_game = False
    
    
def win(board):
    global current_game
    if row(board):
        game_board(board)
        print(f"The winner is {winner}")
        current_game = False

    elif column(board):
        game_board(board)
        print(f"The winner is {winner}")
        current_game = False

    elif diag(board):
        game_board(board)
        print(f"The winner is {winner}")
        current_game = False
        


def player_switcher():
    global current_player
    if current_player == "X":
        current_player = "O"
    else:
        current_player = "X"    


while current_game:
    game_board(board)
    player_input(board)
    player_switcher()
    row(board)
    column(board)
    diag(board)
    tie(board)
    win(board)
```
