```python
import random

##Create a deck with the given suits and ranks

def create_deck():
    suits = ["Spades","Hearts","Diamonds","Clubs"]
    ranks = ["2","3","4","5","6","7","8","9","10","Jack","Queen","King","Ace"]
    
    deck = []
    
    #it sort the cards in the deck in the form of a list
    for suit in suits:
        for rank in ranks:
            deck.append([suit,rank])
            
    return deck

#eg = [Spades,Two] and [Diamonds,One]


## Created a function to transfer money from the loser's acc to the winners acc and substract the given amount from the losers acc
def transfer_money(winner, loser, amount):
    winner['money'] += amount 
    loser['money'] -= amount
    

#INdexing  every rank to find the winner 
def final_winner(card1, card2):
    ranks = ["2", "3", "4", "5", "6", "7", "8", "9", "10", "Jack", "Queen", "King", "Ace"]
    
    rank_one = card1[1]
    rank_two = card2[1]
    
    ##this only takes the last part of the list made in the create_deck function and compares it's index position in the list ranks
    #eg [Spades,Two] , It takes 'Two' and compares it's index position with the card the other player pulled
    
    if ranks.index(rank_one) > ranks.index(rank_two):
        return "Player One"
    elif ranks.index(rank_two) > ranks.index(rank_one):
        return "Player Two"
    else:
        return "No one won"
    
    ##The higher the index position the more the power of the card, so the winner is decided



## This is where it all comes together
def main():
    #the cards are assembled and shuffled
    deck = create_deck()
    random.shuffle(deck)
    
    #each player stats with 1000 credits in their bank acc
    player_one = {'name':'Player One', 'money':1000}
    player_two = {'name':'Player Two', 'money':1000}
    
    #the game goes on until only two cards remain
    while len(deck) >= 2:
        
        print("Player One has:", player_one['money'])
        print("Player Two has:", player_two['money'])
        
        #this takes the bet of the player if they have the money in their account 
        bet_one = int(input("Player one's bet: "))
        bet_two = int(input("Player two's bet: "))
        
        #you can't bet more mony than what is in your acc.
        if bet_one > player_one['money'] or bet_two > player_two['money']:
            print("Invalid bet! Please bet within your capability.")
            continue
        
        print()
        
        #subtracts the betted amount form theaccount
        player_one['money'] -= bet_one
        player_two['money'] -= bet_two
        
        #distributes the card
        card_one = deck.pop(0)  # Draw a card for Player One
        card_two = deck.pop(0)  # Draw a card for Player Two
        
        winner = final_winner(card_one, card_two)
        
        #the winner has the losers money added to their acc and the losers money is subtracted from the acc
        if winner == "Player One":
            transfer_money(player_one, player_two, bet_one + bet_two)
        else:
            transfer_money(player_two, player_one, bet_one + bet_two)
            
    
        #shows the card drawm
        print("Player one draws the card:", card_one)
        print("Player two draws the card:", card_two)
        print(winner, "wins")
        
        #If the money is less than zero than game over
        if player_one['money'] <= 0:
            print("Player One has no money. Game over!")
            break
        elif player_two['money'] <= 0:
            print("Player Two has no money. Game over!")
            break
        
        print()
        
        input("Press enter to play again\n")
        
    print("Deck empty. Goodbye")
    

main()
```
